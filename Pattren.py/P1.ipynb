```python
#--- Pattern 1 ---
# ---i (Job to control rows) is responible for rows and j (Job to print stars) is responsible for columns ---
# --- end="" will not or prevent the j to change the line
# --- print(" ") let i to change the line 

for i in range (1,9+1,1): # outter loop for rows
    for j in range (1,9+1,1): # inner loop for coloumns
        print("* ",end="")
    print(" ")
```
