```python
n=4
x=1 # The number will get incremented by one
for i in range(1,n+1,1):
    for j in range (1,i+1,1): # taking X= i in loop 
        print(x ,end=" ")
        x=x+1
    print(" ")
```

```python
n=4

for i in range(1,n+1,1):
    for j in range(1,i-1+1,1):
        print("  ",end="")
    for j in range(1,(n-i)+(n-i)+1+1,1):
        print("* ",end="")
    print(" ")
```
