```python
#--- Pattern 1 ---
# ---i (Job to control rows) is responible for rows and j (Job to print stars) is responsible for columns ---
# --- end="" will not or prevent the j to change the line
# --- print(" ") let i to change the line 

for i in range (1,9+1,1): # outter loop for rows
    for j in range (1,9+1,1): # inner loop for coloumns
        print("* ",end="")
    print(" ")
```

```python
#--- Pattern 2 ---
# Increasing order of stars in each row

for i in range ( 1,5+1,1): # outter loop for rows
     # if it i increasing order i+/- Something  is used 
     # inner loop for coloumns
     for j in range (1,i+1,1): # Other condition for loop are (i+1),(i-1),
          print("* ",end="") 
     print(" ")
```

```python
#--- Pattern 3 ---
# Decreasing order of stars in each row
n=5
for i in range (1,n+1,1): # outter loop for rows
 # if it is decreasing oreder n-i+1 is used 
    for j in range(1,n-i+1,1): # Other condition for loop are (n-i+1),(n-i-1),(n-i)
        print("* ",end="")
    print(" ")
```
