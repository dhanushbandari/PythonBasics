```python
# Boundary Pattern

n=9

for i in range (1,n+1,1): # 1 to 9 Rows
    for j in range (1,n+1,1): # 1 to 9 Columns
        if i==1 or i==9 or j==1 or j==9: # 1st Row, 9th Row, 1st Column, 9th Column
            print("* ",end="") 
        else:
            print("  ",end="")
    print(" ")

```

```python
#Diagonal Pattern

n=9
for i in range (1,n+1,1): # 1 to 9 Rows
    for j in range(1,n+1,1): # 1 to 9 Columns
        if i==j or i+j==n+1: # 1st Diagonal, 2nd Diagonal
            print("* ",end="")
        else:
            print("  ",end="")
    print(" ")
```

```python
#Pyramid Pattern
n=4
for i in range (1,3+1,1): # 1 to 3 Rows 
    for j in range (1,n-i+1,1): # 1 to 4-1=3, 1 to 4-2=2, 1 to 4-3=1
        print("  ",end="")
    for j in range(1,i+1,1): # 1 to 1, 1 to 2, 1 to 3
        print("* ",end="")
    for j in range (1,i-1+1,1): # 1 to 1-1=0, 1 to 2-1=1, 1 to 3-1=2
        print("* ",end="")
    print("")
# Inverted Pyramid Pattern
n=3
for i in range(1,n+1,1): # 1 to 3 Rows
    for j in range(1,i+1,1): # 1 to 1, 1 to 2, 1 to 3
        print("  ",end="")
    for j in range(1,(n-i)+1,1): # 1 to 3-1=2, 1 to 3-2=1, 1 to 3-3=0
        print("* ",end="")
    for j in range(1,(n-i)+1+1,1): # 1 to 3-1=2, 1 to 3-2=1, 1 to 3-3=0
        print("* ",end="")
    print("")
```
